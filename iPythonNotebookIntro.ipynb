Suppose you want to solve some scientific problem; where do you start?  
For the most part, unless you've solved the problem previously, you  
really don't know *exactly* where; that's the point of a good problem  
---you learn something new as a result of solving it. 

Given that you don't know *exactly* where to start, let's assume that  
you're reasonably intelligent and you have at least *some* idea what  
to do---then you likely won't set out to write one long piece of computer  
code to solve the problem in one go---rather, you'll start to think about  
smaller subpieces of the problem and then glue things together in the end. 

Most scientists work this way---at least initially---when trying to solve  
some problem. You want the ability to be access data, manipulate it, plot it,  
fit it to some model, etc. IPython Notebooks allow all of this to occur  
within an html browser such as Chrome or Firefox or Safari. Let's see what  
iPython Notebooks can do.

<h3> And it's easy to plot functions</h3>

This first cell redefines the default figure size for this session,  
and the second plots a simple sine wave: $ y= \sin (x)$\.

```python
%matplotlib inline
from matplotlib import rcParams
from matplotlib import rc
rc('text', usetex=True)    # this uses LaTeX fonts to typeset figures
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['font.size'] = 16
import matplotlib as mpl
import matplotlib.pylab as pylab  # imports matplotlib plotting package pylab
import matplotlib.pyplot as plt   # imports matplotlin plotting package pyplot
pylab.rcParams['figure.figsize'] = 8, 6
```

If the following cell doesn't work and gives an error message 

! LaTeX Error: File `type1cm.sty' not found.

then you need to make sure that you have texlive-latex-extra  
installed with your (linux) texlive installation. On MacOS,  
this should have been installed when you download macTeX;  
on linux, just open a BASH terminal and type:

sudo apt-get install texlive-latex-extra
sudo apt-get install dvipng

and re-evaluate the next cell, and you should be all set.

```python
import numpy as np
t=np.linspace(0,4*np.pi,200) # creates a 200 element 1D array from [0,4*pi] 
plt.xlabel(r"time (s)", size=20)
plt.ylabel(r"Electric Field (N/C)", size = 20)
plt.plot(t,np.sin(t))
plt.savefig("filename.pdf", bbox_inches='tight', dpi=300)
```

Now here's the same plot with  slightly larger tick labels, boldfaced axes labels, and a title composed in $\LaTeX$.

```python
t=np.linspace(0,4*np.pi,200) # creates a 200 element 1D array from [0,4*pi] 
plt.xlabel(r'\textbf{time (s)}', size=20) 
plt.ylabel(r"\textbf{Electric Field (N/C)}", size = 20)
plt.title(r"\TeX\ is Number $\displaystyle\sum_{n=1}^\infty\frac{-e^{i\pi}}{2^n}$!", fontsize=20, color='r')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.tick_params(axis='both', which='minor', labelsize=12)

plt.plot(t,np.sin(t))
plt.savefig("filename.pdf", bbox_inches='tight', dpi=300)
```

<h3>Of course, if you have a set of data with error bars, you can plot that data too:</h3>

```python
t = np.linspace(0.2,2,10)
temp = np.array([8.24,13.76,17.47, 19.95,21.62, 22.73, 23.48, 23.98, 24.32, 24.54])
dt = np.array([0.9,0.8,0.7,0.6,0.5, 0.4, 0.3, 0.2, 0.1,0.1])
print 'times (s)= ', t
print 'temperature (C) = ',temp
print '∆t (C) = ', dt
plt.xlabel(r'\textbf{time (s)}', size=20) 
plt.ylabel(r"\textbf{Temperature (C)}", size = 20)
plt.ylim(0,25)

plt.errorbar(t,temp,yerr=dt,marker='.',color='b',linestyle='none')
```

<h3>And you can easily include python code and $\LaTeX$ mathematics</h3>

If you create a markdown cell, you can include python code by simply  
skipping a line, and indenting, and the iPython notebook automatically  
detects what language is present, in this case some python code:

    from pylab import *
    t = linspace(0,10,200)
    plot(t,sin(t))
    show()

copy this code and paste it into a new cell below this one, and see that it runs!

In addition, you can currently do CRUDE linebreaking manually by putting two spaces  
followed by a return; markdown automatically converts that to a line break. Somewhat 
awkward, but on the bleeding edge, that's what you have to be content with. 

Markdown cells can also display $\LaTeX$ code either inline: $\sin(2\pi) = 0$, or as

$$ \int_0^{2} x^2\;dx = \frac{8}{3}$$
